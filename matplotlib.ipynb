## Scipy:
## We have the min and max temperatures in a city In India for each months of the year.
## We would like to find a function to describe this and show it graphically, the dataset given below.

Task:
1. fitting it to the periodic function
2. plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18
m = np.arange(12)
```

```python
def periodic_func(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))
```

```python
res_max, cov_max = optimize.curve_fit(periodic_func, m, Max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(periodic_func, m, Min, [-40, 20, 0])
```

```python
d = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(m, Max, 'ro')
plt.plot(d, periodic_func(d, *res_max), 'r-')
plt.plot(m, Min, 'bo')
plt.plot(d, periodic_func(d, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```
