# In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
x=iris.iloc[:,:-1]
x.head()
```

```python
y=iris.iloc[:,-1]
y.head()
```

```python
pca = decomposition.PCA(n_components=3)
pca.fit(x)
x = pca.transform(x)
```

```python
fig = plt.figure(1, figsize=(10, 10))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)

plt.cla()

# Reorder the labels to have colors matching the cluster results
ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=y, cmap=plt.cm.get_cmap("Spectral"), edgecolor='k')
plt.show()
```
